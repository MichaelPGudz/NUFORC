```python
import pandas as pd
import requests
import time
import random
from datetime import datetime
from bs4 import BeautifulSoup
from urllib.parse import urljoin
```

```python
def get_monthly_event_pages(events_url='http://www.nuforc.org/webreports/ndxevent.html'):
    page = requests.get(events_url)
    soup = BeautifulSoup(page.text, 'html.parser')
    monthly_event_pages = [urljoin('http://www.nuforc.org/webreports/', x['href']) for x in list(soup.find_all('a', href=True))[1:-1]]
    return monthly_event_pages

def scrape_monthly_event_page(url):
    keys = ['datetime', 'city', 'state', 'shape', 'duration', 'summary', 'posted'] 
    
    page = requests.get(url)
    soup = BeautifulSoup(page.text, 'html.parser')
    monthly_event_tags = soup.find_all('tr')[1:]
    monthly_event_list = [dict(zip(keys, tag.text.split('\n')[1:-1])) for tag in monthly_event_tags]
    
    return monthly_event_list
```

```python
monthly_event_pages = get_monthly_event_pages()
```

```python
all_events = []

start_time = datetime.now()
for i, url in enumerate(monthly_event_pages):
    print("Scraping URL {}/{}: {}".format(i+1, len(monthly_event_pages), url))
    monthly_event_list = scrape_monthly_event_page(url)
    all_events.extend(monthly_event_list)
    time.sleep(random.randint(5, 10))
end_time = datetime.now()
print('Duration: {}'.format(end_time - start_time))
```

```python
pd.DataFrame(all_events).to_csv('data/raw_nuforc.csv', index=False)
```
